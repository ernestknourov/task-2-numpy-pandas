**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
df = data[data.salary == '>50K'] # data about person with big salary

total_married = data[data['marital-status'].str.startswith('Married')]['marital-status'].shape[0]
rich_married = df[df['marital-status'].str.startswith('Married')]['marital-status'].shape[0]

total_single = data[~data['marital-status'].str.startswith('Married')]['marital-status'].shape[0]
rich_single = df[~df['marital-status'].str.startswith('Married')]['marital-status'].shape[0]

print(rich_married/total_married)
print(rich_single/total_single)
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours = data['hours-per-week'].max()
persons = data[data['hours-per-week'] == max_hours]
print(f"Maximum hours-per-week: {max_hours}")
print(f"Number of person working as much time: {persons.shape[0]}")
print(f"Percent of person with big salary: {100*persons[persons.salary == '>50K'].shape[0]/persons.shape[0]}")
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
# 1st way, less optimized
# fun = lambda x: "young" if x >=16 and x <35 else ("adult" if x >= 35 and x < 70 else "retiree")
# AgeGroup = [fun(x) for x in data['age']]

# 2nd way
AgeGroup = pd.cut(data["age"],[16,35,70,100],labels=["young","adult","retiree"])
data.insert(1, "AgeGroup", AgeGroup)
data
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
df = data[data.salary == '>50K']
print(df.groupby(['AgeGroup'])['AgeGroup'].count())
per_young = df[df.AgeGroup == 'young'].shape[0]/data[data.AgeGroup == 'young'].shape[0]*100 # percent of rich young (salary>50K)
per_adult = df[df.AgeGroup == 'adult'].shape[0]/data[data.AgeGroup == 'adult'].shape[0]*100 # percent of rich adult (salary>50K)
per_retiree = df[df.AgeGroup =='retiree'].shape[0]/data[data.AgeGroup == 'retiree'].shape[0]*100 # percent of rich retiree (salary>50K)

print(30*'-')
print(per_young)
print(per_adult)
print(per_retiree)
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
group_data = data.groupby('occupation')
group_data['occupation'].count()
```

```python
def filter_func(df):
    return (df['age'].mean() < 40) & (df['hours-per-week'].min() > 5)

group_data.filter(filter_func)
```
